# KMeans

```python
kmeans= KMeans(n_clusters=10, random_state = 0 )
kmeans.fit(final_norm)
```

```python
kmeans.inertia_
```

```python
from sklearn.metrics import silhouette_score, calinski_harabasz_score 

wcss = [] 
sil_score = []  
cal_haz_score = [] 

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(final_norm)
    wcss.append(kmeans.inertia_)
    if i==1:
        continue 
    sil_score.append(silhouette_score(final_norm, kmeans.labels_))
    cal_haz_score.append(calinski_harabasz_score(final_norm, kmeans.labels_))
```

```python
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
